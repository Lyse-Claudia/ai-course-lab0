```python
"""
Exercise 5: Pandas Data Analysis
Analyze a dataset of student performance.
"""

import pandas as pd
import numpy as np

# Create sample dataset
np.random.seed(42)
n_students = 200

data = {
    'student_id': range(1000, 1000 + n_students),
    'major': np.random.choice(['CS', 'Math', 'Physics', 'Biology'], n_students),
    'year': np.random.choice([1, 2, 3, 4], n_students),
    'exam_score': np.random.normal(75, 10, n_students).clip(0, 100),
    'assignments_completed': np.random.randint(0, 11, n_students),
    'hours_studied': np.random.normal(15, 5, n_students).clip(1, 40)
}

df = pd.DataFrame(data)
print(df)
# Introduce some NaN values
df.loc[np.random.choice(n_students, 10), 'exam_score'] = np.nan
df.loc[np.random.choice(n_students, 5), 'hours_studied'] = np.nan

print(df)
# Task 1: Data Cleaning and Exploration (10 points)

print(df.head(2))
print(df.info())
print(df.describe())
print(df.shape)

# Identify and count missing values
print(df.isnull().sum())

# Fill missing exam_score with the mean score for the student's major
df['exam_score']= df['exam_score'].fillna(df.groupby('major')['exam_score'].transform('mean'))

# Fill missing hours_studied with the median for the student's year
df['hours_studied']=df['hours_studied'].fillna(df.groupby('year')['hours_studied'].transform('median'), inplace=True)
print(df)

# Task 2: Analysis (10 points)
# Calculate and display the average exam_score by major

average_by_major = df.groupby('major')['exam_score'].mean()
print(average_by_major)

# Find the major with the highest average exam_score
highest_major = average_by_major.idxmax()
print('Major with highest average exam score: ',highest_major)

# Calculate the correlation between hours_studied and exam_score
correlation = df['hours_studied'].corr(df['exam_score'])
print('Correlation between hours studied and exam score: ',correlation)

# TOCrea te a new column 'performance' with categories:
#       'Excellent' (>90), 'Good' (80-90), 'Average' (70-80), 'Needs Improvement' (<70)
def categorize_performance(score):
    if score > 90:
        return 'Excellent'
    elif score > 80:
        return 'Good'
    elif score > 70:
        return 'Average'
    else:
        return 'Needs Improvement'
df['performance'] = df['exam_score'].apply(categorize_performance)
print(df[['exam_score', 'performance']].head(10))   


# Task 3: Advanced Analysis (10 points)

grouped = df.groupby(['major', 'year']).agg(
    num_students=('student_id', 'count'),
    avg_exam_score=('exam_score', 'mean'),
    avg_hours_studied=('hours_studied', 'mean')
).reset_index()
print(grouped)

# Identify top 5 students based on exam_score (handle ties appropriately)
top5 = df.nlargest(5, 'exam_score')
print(top5[['student_id', 'exam_score']])

# Create a pivot table showing average exam_score by major (rows) and year (columns)
pivot_table = df.pivot_table(values='exam_score', index='major', columns='year', aggfunc='mean')
print(pivot_table)
```

```python
"""
Exercise 6: Data Visualization
Create meaningful visualizations using the dataset from Exercise 5.
"""

import matplotlib.pyplot as plt
import seaborn as sns

# Continue using the df from Exercise 5

# Task 1: Distribution Visualization (8 points)

fig,ax = plt.subplots(1,2,figsize=(12,5))
ax1 = ax[0]
ax2 = ax[1]

sns.histplot(df['exam_score'],kde=True, bins=20, ax=ax1)
ax1.set_title("Distribution of Exam Scores")
ax1.set_xlabel("Exam Score")
ax1.set_ylabel("Frequency")

sns.boxplot(x='major', y='exam_score', data=df,ax=ax2)
ax2.set_title('Exam Scores by Major')
ax2.set_xlabel('Major')
ax2.set_ylabel('Exam Score')

plt.suptitle('Exam Score Analysis', fontsize=14)
plt.tight_layout()
plt.show()

# Task 2: Relationship Visualization (8 points)

fig, ax = plt.subplots(figsize=(10,6))
sns.scatterplot(x='hours_studied', y='exam_score', hue='major', ax=ax)
sns.regplot(x='hours_studied', y='exam_score', data=df, scatter=False, color='black', ax=ax)
ax.set_title('Hours Studied vs Exam Score')
ax.set_xlabel('Hours Studied')
ax.set_ylabel('Exam Score')

plt.tight_layout()
plt.show()


# Task 3: Advanced Dashboard (9 points)

fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2,figsize=(15,12))
average_scores= df.groupby('major')['exam_score'].mean()
ax1.bar(average_scores.index, average_scores.values)
ax1.set_title('Average Exam Score by Major')
ax1.set_xlabel('Major')
ax1.set_ylabel('Average Score')

sns.countplot(data=df, x='year', ax=ax2)
ax2.set_title('Number of Students by Year')
ax2.set_xlabel('Year')
ax2.set_ylabel('Count')

numerical_cols = ['exam_score', 'assignments_completed', 'hours_studied']
sns.heatmap(df[numerical_cols].corr(), annot=True, fmt='.2f', cmap='coolwarm', ax=ax3)
ax3.set_title('Correlation Matrix')

sns.violinplot(data=df, x='performance', y='exam_score', ax=ax4)
ax4.set_title('Exam Score by Performance Category')
ax4.set_xlabel('Performance')
ax4.set_ylabel('Exam Score')

plt.suptitle('Student Performance Dashboard', fontsize=16)
plt.tight_layout()  
plt.show()
```

```python
"""
Exercise 7: Integration Challenge
Combine NumPy, Pandas, and Matplotlib to solve a mini data science problem.
"""

# Scenario: You're analyzing customer data for an e-commerce company.
# Generate synthetic data and provide insights.

np.random.seed(42)
n_customers = 500

# Generate customer data
ages = np.random.randint(18, 70, n_customers)
income = np.random.normal(50000, 20000, n_customers).clip(15000, 150000)
purchase_freq = np.random.poisson(5, n_customers)
avg_purchase_value = np.random.normal(100, 30, n_customers).clip(10, 500)

# Create DataFrame
customers = pd.DataFrame({
    'age': ages,
    'income': income,
    'purchase_frequency': purchase_freq,
    'avg_purchase_value': avg_purchase_value
})

max_frequency = customers['purchase_frequency'].max()
customers['churn_risk'] = 1 - (customers['purchase_frequency'] / max_frequency)
customers['CLV'] = customers['purchase_frequency'] * customers['avg_purchase_value'] * (1 + customers['churn_risk'])

# Create age groups: 18-25, 26-35, 36-50, 51-70
def age_group(age):
    if age <= 25:
        return '18-25'
    elif age <= 35:
        return '26-35'
    elif age <= 50:
        return '36-50'
    else:
        return '51-70'
customers['age_group'] = customers['age'].apply(age_group)  

age_group_summary = customers.groupby('age_group').agg(
    num_customers=('age', 'count'),
    avg_income=('income', 'mean'),
    avg_CLV=('CLV', 'mean'),
    total_CLV=('CLV', 'sum')
).reset_index() 
print(age_group_summary)


# Identify top 10% of customers by CLV
top_10 = customers['CLV'].quantile(0.9)
top_customers = customers[customers['CLV'] >= top_10]


fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 5))

# 1. Scatter plot: income vs CLV colored by age group
sns.scatterplot(data=customers, x='income', y='CLV', hue='age_group', alpha=0.6, ax=ax1)
ax1.set_title('Income vs CLV by Age Group')
ax1.set_xlabel('Income')
ax1.set_ylabel('CLV')

# 2. Bar chart: average CLV by age group
avg_clv = customers.groupby('age_group')['CLV'].mean()
ax2.bar(avg_clv.index, avg_clv.values)
ax2.set_title('Average CLV by Age Group')
ax2.set_xlabel('Age Group')
ax2.set_ylabel('Average CLV')

# 3. Correlation heatmap
numerical_cols = ['income', 'CLV', 'age']
sns.heatmap(customers[numerical_cols].corr(), annot=True, fmt='.2f', cmap='coolwarm', ax=ax3)
ax3.set_title('Correlation Matrix')

plt.suptitle('Customer Lifetime Value Analysis', fontsize=16)
plt.tight_layout()
plt.show()

```

## Summary

The data shows that purchase frequency is the strongest driver of CLV. In addition, 26-50 age group have the highest average CLV. Therefore, they should prioritize retaining that age group and develop strategies for to gain trust from them .
